```python
from tqdm import tqdm
```

```python
# Define the paths to the English and Spanish folders
english_folder = '/kaggle/input/twitter-dataset-for-mental-disorders-detection/Twitter dataset/Timelines/English'
spanish_folder = '/kaggle/input/twitter-dataset-for-mental-disorders-detection/Twitter dataset/Timelines/Spanish'
```

```python
# Create empty DataFrames for English and Spanish data
english_data = pd.DataFrame()
spanish_data = pd.DataFrame()

# Function to merge all CSV files in a folder
def merge_csv(folder_path, dataframe):
    for root, dirs, files in tqdm(os.walk(folder_path)):
        for file in tqdm(files):
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                df = pd.read_csv(file_path)
                dataframe = pd.concat([dataframe, df], ignore_index=True)
    return dataframe
```

```python
# Merge all English CSV files
english_data = merge_csv(english_folder, english_data)

# Save the merged data to new CSV files
english_data.to_csv('/kaggle/working/merged_english_tweets.csv', index=False)

print("Merging completed and files saved.")
```

```python
# Merge all Spanish CSV files
spanish_data = merge_csv(spanish_folder, spanish_data)
spanish_data.to_csv('/kaggle/working/merged_spanish_tweets.csv', index=False)
print("Merging completed and files saved.")
```
